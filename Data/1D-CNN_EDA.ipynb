```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python
# Load the dataset
data = pd.read_csv('Data/heart.csv')

# Display the first few rows
data.head()

# Check the shape of the dataset
print(f"Dataset shape: {data.shape}")

# Check the data types and non-null counts
data.info()
```

```python
# Display summary statistics of numeric columns
data.describe()
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
```

```python
# Distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x=data['target'], palette="viridis")
plt.title('Distribution of Target Variable (Heart Disease)')
plt.xlabel('Target (0: No Disease, 1: Disease)')
plt.ylabel('Count')
plt.show()
```

```python
# Plot histograms for all numeric features
data.hist(figsize=(15, 12), bins=15, color='steelblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = data.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplot for 'cp' (chest pain) vs 'target'
plt.figure(figsize=(8, 6))
sns.boxplot(x='cp', y='target', data=data, palette="viridis")
plt.title('Chest Pain Type vs Target')
plt.show()

# Boxplot for 'thal' vs 'target'
plt.figure(figsize=(8, 6))
sns.boxplot(x='thal', y='target', data=data, palette="viridis")
plt.title('Thalassemia vs Target')
plt.show()
```

```python
# Visualizing relationships between features and the target

# Plot distribution of 'age' based on 'target'
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='age', data=data, palette="viridis")
plt.title('Age vs Target')
plt.show()

# Plot distribution of 'chol' (cholesterol) based on 'target'
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='chol', data=data, palette="viridis")
plt.title('Cholesterol vs Target')
plt.show()
```
